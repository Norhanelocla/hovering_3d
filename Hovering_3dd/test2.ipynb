```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib.animation import FuncAnimation
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
t = np.linspace(0, 10, 100)
drone_x = np.sin(t)
drone_y = np.cos(t)
drone_z = t
ax = fig.add_subplot(111, projection='3d')
window_x = np.sin(t) + 2  # Offset the second trajectory in X
window_y = np.cos(t) + 2  # Offset the second trajectory in Y
window_z = t
scatter, = ax.plot(window_x, window_y, window_z)
scatter2, = ax.plot(drone_x, drone_y,drone_z)
plt.show()
```

```python
scattter = []
scattter2 =[]
def update(num, x, y,z,x2,y2,z2, scattter,scattter2):
    scattter = ax.scatter(x2[:num], y2[:num], z2[:num], c=z2[:num], cmap='plasma')
    scattter2 = ax.scatter(x[:num], y[:num], z[:num], c=z[:num], cmap='viridis')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    return scattter,scattter2

ani = FuncAnimation(fig, update, frames=range(0, len(drone_x), 100), fargs=[drone_x, drone_y, drone_z, window_x, window_y, window_z, scattter, scattter2], interval=1000)
from IPython.display import HTML

HTML(ani.to_html5_video())
```
